```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mplt
import seaborn as sb
```

```python
#loading the dataset
df = pd.read_csv('retail_sales_dataset.csv') 
```

```python
print("DATASET PREVIEW")
df.head()
```

```python
# Checking for missing values and dataset summary
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
print("\nDataset Information:")
df.info()
```

```python
# Gender distribution
mplt.figure(figsize=(6, 4))
sb.countplot(data=df, x='Gender', hue='Gender')  
mplt.title('Gender Distribution')
mplt.show()
```

```python
# Age distribution
mplt.figure(figsize=(8, 5))
sb.histplot(data=df, x='Age', kde=True, bins=15)  
mplt.title('Age Distribution')
mplt.show()
```

```python
# Product Category distribution
mplt.figure(figsize=(10, 6))
sb.countplot(data=df, x='Product Category', hue='Product Category')  
mplt.title('Sales Distribution by Product Category')
mplt.xticks(rotation=45)
mplt.show()
```

```python
# Total sales per product category
mplt.figure(figsize=(10, 6))
sb.barplot(data=df, x='Product Category', y='Total Amount', estimator=sum) 
mplt.title('Total Sales per Product Category')
mplt.xticks(rotation=45)
mplt.show()
```

```python
# Sales trend over time
mplt.figure(figsize=(12, 6))
df.groupby('Date')['Total Amount'].sum().plot() 
mplt.title('Total Sales Trend Over Time')
mplt.xlabel('Date')
mplt.ylabel('Total Sales')
mplt.show()
```

```python
# Scatter plot for Price vs. Quantity
mplt.figure(figsize=(8, 5))
sb.scatterplot(data=df, x='Price per Unit', y='Quantity', hue='Product Category')
mplt.title('Price vs. Quantity by Product Category')
mplt.show()
```

```python
# Boxplot for total amount by gender
mplt.figure(figsize=(8, 6))
sb.boxplot(data=df, x='Gender', y='Total Amount', )  
mplt.title('Boxplot of Total Amount by Gender')
mplt.show()
```
